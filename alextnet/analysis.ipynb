```python
import pickle

# load alexnet_cifar10_history.pkl
with open('alexnet_cifar10_history.pkl', 'rb') as f1:
    alexnet_history = pickle.load(f1)

with open('modern_alexnet_cifar10_history.pkl', 'rb') as f2:
    modern_alexnet_history = pickle.load(f2)
    
with open('modern_alexnet_adamW_cifar10_history.pkl', 'rb') as f3:
    modern_alexnet_adamW_history = pickle.load(f3)
```

```python
import matplotlib.pyplot as plt

alexnet_val_acc = alexnet_history['val_accuracy']
modern_alexnet_val_acc = modern_alexnet_history['val_accuracy']
modern_alexnet_adamW_val_acc = modern_alexnet_adamW_history['val_accuracy']
alexnet_val_loss = alexnet_history['loss']
modern_alexnet_val_loss = modern_alexnet_history['val_loss']
modern_alexnet_adamW_val_loss = modern_alexnet_adamW_history['val_loss']

epochs_range = range(len(alexnet_val_acc))

# Plot Training and Validation Accuracy
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, alexnet_val_acc, label='AlexNet Validation Accuracy')
plt.plot(epochs_range, modern_alexnet_val_acc, label='Modern AlexNet Validation Accuracy')
plt.plot(epochs_range, modern_alexnet_adamW_val_acc, label='Modern AlexNet AdamW Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Validation Accuracy Comparison')

# Plot Training and Validation Loss
plt.subplot(1, 2, 2)
plt.plot(epochs_range, alexnet_val_loss, label='AlexNet Training Loss')
plt.plot(epochs_range, modern_alexnet_val_loss, label='Modern AlexNet Training Loss')
plt.plot(epochs_range, modern_alexnet_adamW_val_loss, label='Modern AlexNet AdamW Training Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.show()
```
